```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
def get_mid(a, b):
    x = (a[0] + b[0]) / 2
    y = (a[1] + b[1]) / 2
    return [x, y]
```

```python
def draw_san(size, i):
    sanjiaoxing(size)
    if i > 0:
                 # Нарисуйте маленький треугольник слева
        size2 = [size[0], get_mid(size[0], size[1]), get_mid(size[0], size[2])]
        draw_san(size2, i - 1)
 
                 # Нарисуйте верхний треугольник
        size3 = [get_mid(size[0], size[2]), get_mid(size[1], size[2]), size[2]]
        draw_san(size3, i - 1)
        
                 # Нарисуйте маленький треугольник справа
        size4 = [get_mid(size[0], size[1]), size[1], get_mid(size[1], size[2])]
        draw_san(size4, i - 1)
```

```python
def sanjiaoxing(san):
  x = []
  y = []
  for coord in san:
    x.append(coord[0])
    y.append(coord[1])
  x.append(san[0][0])
  y.append(san[0][1])

  plt.plot(x, y, '-k')
```

```python
points = [[-200, 0], [200, 0], [0, 300]]
      # Рекурсия count раз
count = 4
draw_san(points, count);
 
```
